```python
plt.figure(figsize=(12, 5))

# Plotting the first subplot (young working-age population)
plt.subplot(1, 2, 1)
sns.barplot(x='Country', y='% of young working-age population- 2022', data=df.head(20), palette='viridis')
plt.xlabel('Country')
plt.ylabel('% of young working-age population- 2022')
plt.title('Distribution of young working age population %')
plt.xticks(rotation=80)

# Plotting the second subplot (old working-age population)
plt.subplot(1, 2, 2)
sns.barplot(x='Country', y='% of old working-age population- 2022', data=df.head(20), palette='viridis')
plt.xlabel('Country')
plt.ylabel('% of old working-age population- 2022')
plt.title('Distribution of old working age population %')
plt.xticks(rotation=80)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
